```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import time
import csv
from collections import defaultdict
import pandas as pd
from pandas import DataFrame as df
from selenium.webdriver.support.wait import WebDriverWait
from urllib.request import urlopen
import lxml



    
data = []
url = 'https://www.basketball-reference.com/leagues/NBA_2003_totals.html' 
soup = BeautifulSoup(urlopen(url),'lxml')
soup.prettify()
tr = soup.find_all('tr',class_ = 'full_table')
for info in tr:
    result = {}
    result['Player_Name'] = info.find('td', class_ = 'left').text
    result['Position'] = info.find('td', class_ = 'center').text    
    result['Age'] = info.find('td',class_='right').text
    result['Team'] = info.find('td', attrs={'data-stat' : 'team_id'}).text
    result['Games'] = info.find('td', attrs={'data-stat' : 'g'}).text
    result['Games_Strated'] = info.find('td', attrs={'data-stat' : 'gs'}).text
    result['Minutes_Played'] = info.find('td', attrs={'data-stat' : 'mp'}).text
    result['FG'] = info.find('td', attrs={'data-stat' : 'fg'}).text
    result['FGA'] = info.find('td', attrs={'data-stat' : 'fga'}).text
    result['FG%'] = info.find('td', attrs={'data-stat' : 'fg_pct'}).text
    result['3P'] = info.find('td', attrs={'data-stat' : 'fg3'}).text
    result['3PA'] = info.find('td', attrs={'data-stat' : 'fg3a'}).text
    result['3P%'] = info.find('td', attrs={'data-stat' : 'fg3_pct'}).text
    result['2P'] = info.find('td', attrs={'data-stat' : 'fg2'}).text
    result['2PA'] = info.find('td', attrs={'data-stat' : 'fg2a'}).text
    result['2P%'] = info.find('td', attrs={'data-stat' : 'fg2_pct'}).text
    result['eFG%'] = info.find('td', attrs={'data-stat' : 'efg_pct'}).text
    result['FT'] = info.find('td', attrs={'data-stat' : 'ft'}).text
    result['FTA'] = info.find('td', attrs={'data-stat' : 'fta'}).text
    result['FT%'] = info.find('td', attrs={'data-stat' : 'ft_pct'}).text
    result['ORB'] = info.find('td', attrs={'data-stat' : 'orb'}).text
    result['DRB'] = info.find('td', attrs={'data-stat' : 'drb'}).text
    result['TRB'] = info.find('td', attrs={'data-stat' : 'trb'}).text
    result['AST'] = info.find('td', attrs={'data-stat' : 'ast'}).text
    result['STL'] = info.find('td', attrs={'data-stat' : 'stl'}).text
    result['BLK'] = info.find('td', attrs={'data-stat' : 'blk'}).text
    result['TOV'] = info.find('td', attrs={'data-stat' : 'tov'}).text
    result['PF'] = info.find('td', attrs={'data-stat' : 'pf'}).text
    result['PTS'] = info.find('td', attrs={'data-stat' : 'pts'}).text
    data.append(result)
#print(data)
df = pd.DataFrame.from_records(data)
cols = list(df)
cols.insert(0, cols.pop(cols.index('Player_Name')))
df = df.ix[:, cols]
print(df)
df.to_csv('2002-03.csv')
```
